# Solve Me First

> **Platform:** HackerRank  
> **Difficulty:** Easy  
> **Language:** C#

---

## 📖 Problem

Given two integers `a` and `b`, return their sum.

---

## Idea

The problem provides exactly **two integers**.

Since the number of inputs is fixed, we don't need loops or advanced algorithms.

The solution is simply to add the two numbers and return the result.

---

## Pattern

### Direct Computation

**Definition**

Use this pattern when the problem requires performing an operation on a **fixed number of values**.

### Characteristics

- No loops.
- No iteration.
- No accumulation.
- Constant execution time.

### Common Use Cases

- Add two numbers.
- Subtract two numbers.
- Compare two values.
- Find the maximum/minimum of two values.

---

## Algorithm

1. Receive the first integer.
2. Receive the second integer.
3. Compute their sum.
4. Return the result.

---

## Complexity Analysis

| Complexity | Value |
|------------|-------|
| **Time Complexity** | **O(1)** |
| **Space Complexity** | **O(1)** |

### Why?

Only one addition operation is performed regardless of the input values.

---

## What I Learned

- How to analyze a problem before writing code.
- How to identify the **Input**, **Output**, and **Constraints**.
- The difference between an **Algorithm** and its **Implementation**.
- Constraints describe the expected input and do not necessarily require validation.
- Introduced the **Direct Computation** pattern.

---

## Common Mistakes

- Adding unnecessary input validation even though the problem doesn't require it.
- Writing a complete Console Application instead of implementing only the required function.
- Confusing **Algorithm** with **Implementation**.

---

## Key Takeaways

- Always understand the problem before writing code.
- Think about the algorithm first, then translate it into code.
- Start with the simplest possible solution before looking for optimizations.
